```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
X_wine_scaled = scaler.fit_transform(X_wine)
X_yoga_scaled = scaler.fit_transform(X_yoga)
```

```python
k = 3
kmeans_wine = KMeans(n_clusters=k, random_state=42)
wine_clusters = kmeans_wine.fit_predict(X_wine_scaled)

kmeans_yoga = KMeans(n_clusters=k, random_state=42)
yoga_clusters = kmeans_yoga.fit_predict(X_yoga_scaled)
```

```python
import numpy as np

print("Wine clusters:", np.bincount(wine_clusters))
print("Yoga clusters:", np.bincount(yoga_clusters))
```

```python
# Wine
plt.figure(figsize=(10,6))
for cluster in range(k):
    mean_series = X_wine_scaled[wine_clusters == cluster].mean(axis=0)
    plt.plot(mean_series, label=f"Cluster {cluster}")
plt.title("   Wine  ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.legend()
plt.show()

# Yoga
plt.figure(figsize=(10,6))
for cluster in range(k):
    mean_series = X_yoga_scaled[yoga_clusters == cluster].mean(axis=0)
    plt.plot(mean_series, label=f"Cluster {cluster}")
plt.title("   Yoga  ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)  # 2D 

wine_pca = pca.fit_transform(X_wine_scaled)
yoga_pca = pca.fit_transform(X_yoga_scaled)
```

```python
# Wine
plt.figure(figsize=(8,6))
for cluster in range(k):
    plt.scatter(wine_pca[wine_clusters==cluster, 0],
                wine_pca[wine_clusters==cluster, 1],
                label=f"Cluster {cluster}")
plt.title("Wine:   (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()

# Yoga
plt.figure(figsize=(8,6))
for cluster in range(k):
    plt.scatter(yoga_pca[yoga_clusters==cluster, 0],
                yoga_pca[yoga_clusters==cluster, 1],
                label=f"Cluster {cluster}")
plt.title("Yoga:   (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()
```

```python
wine_train = pd.read_csv("WineTRAIN", header=None)
wine_test  = pd.read_csv("WineTEST", header=None)

X_train = wine_train.iloc[:, 1:]
y_train = wine_train.iloc[:, 0]

X_test = wine_test.iloc[:, 1:]
y_test = wine_test.iloc[:, 0]

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train_scaled, y_train)

y_pred = clf.predict(X_test_scaled)

```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

```
